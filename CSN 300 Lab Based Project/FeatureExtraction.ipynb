```python
from __future__ import print_function
import sys
import os
import logging
import six
from radiomics import featureextractor
import radiomics
import csv
```

```python
# log file initialisation

logger = radiomics.logger
logger.setLevel(logging.DEBUG)

handler = logging.FileHandler(filename='testLog.txt', mode='w')
formatter = logging.Formatter('%(levelname)s:%(name)s: %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)
```

```python
#code to collect the images of different patients
path = "../MICCAI_BraTS_2018_Data_Training/HGG"
#In the above line use the absolute address when using in your machine

imagesIn=os.listdir(path)
#list containing all the names of different cases
initpath = "/media/joseph/New Volume/Academics/Third yr Spring/Lab Based Project/BrainTumor/BraTS_2018/MICCAI_BraTS_2018_Data_Training/HGG"

#print(initpath + "/" + imagesIn[0]+"_t1.nii.gz")
#print("mask  "+ initpath + "/" + imagesIn[0]+"_seg.nii.gz")
```

##### EXTRACTION OF RADIOMICS FEATURES FROM T1 IMAGES 

```python
j = 0
with open('TrainingDataHGGT1.csv', 'w') as outfile:
        for img in imagesIn :
            imageNameT1 = initpath+"/"+img+"/" +img+"_t1.nii.gz"
            #imageNameT2 = initpath + "/" + img+"_t2.nii.gz"
            #imageNameFLAIR = initpath + "/" + img+"_flair.nii.gz"
            #imageNameT1CE = initpath + "/" + img+"_t1ce.nii.gz"
            maskName = initpath + "/" + img+"/" +img+"_seg.nii.gz"
            if imageNameT1 is None or maskName is None:
                raise Exception('Error getting testcase!')
            #print(imageName)
            params = os.path.join('..', 'Params.yaml')
            #print(params)
            #print(maskName)
            extractor = featureextractor.RadiomicsFeaturesExtractor(params)
            print('Calculating features for', img)
            featureVector = extractor.execute(imageNameT1, maskName)
            if j==0 :
                # adding the name of the patient folder as a fieldname
                f_names = featureVector.keys()
                f_names.insert(len(f_names), "folder_name")
                #fieldnames = featureVector.keys()
                fieldnames = f_names
                writer = csv.DictWriter(outfile, fieldnames= fieldnames)
                writer.writeheader()
            j=1
            #adding 'folder_name' - key and its value to the featureVector dictionary
            feature_dict = featureVector
            new_key = "folder_name"
            new_value = img
            feature_dict[new_key]=new_value
            #print("feature_dict is ", feature_dict)
            #print("featureVector is", featureVector)
            #writer.writerow(featureVector)
            writer.writerow(feature_dict)
#extractor = featureextractor.RadiomicsFeaturesExtractor(params)
#extractor.correctMask=True
#print('Calculating features for', img1)
#featureVector = extractor.execute(img1, maskName)
outfile.close()
```
